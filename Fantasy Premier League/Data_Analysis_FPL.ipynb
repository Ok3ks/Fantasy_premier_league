```python
import numpy as np
import pandas as pd
import requests
```

```python
url = "https://fantasy.premierleague.com/api/bootstrap-static/"

response = requests.get(url)

json = response.json()

json.keys()
```

```python
elements_df = pd.DataFrame(json['elements'])
elements_df.columns
```

```python
events_df = pd.DataFrame(json['events'])
events_df.columns

```

```python
elements_df = pd.DataFrame(json['elements'])
elements_types_df = pd.DataFrame(json['element_types'])
teams_df = pd.DataFrame(json['teams'])

elements_types_df.columns
teams_df
```

```python
slim_elements_df = elements_df[['second_name', 'team', 'element_type', 'selected_by_percent', 'now_cost', 'minutes', 'value_season','transfers_in']]



slim_elements_df.now_cost = slim_elements_df.now_cost/10
slim_elements_df

slim_elements_df['position'] = slim_elements_df.element_type.map(elements_types_df.set_index('id').singular_name)
slim_elements_df['teams'] = slim_elements_df.team.map(teams_df.set_index('id').name)



```
